```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt'
names = ['variance', 'skewness', 'curtosis', 'entropy', 'class']
dataset = pd.read_csv(url, names=names)
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Training the logistic regression model on the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Calculating the Accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

The above code does the following:

Imports the necessary libraries: NumPy, Matplotlib, and Pandas for data manipulation and visualization, and Scikit-Learn for machine learning algorithms.

Loads the banknote authentication dataset from the UCI Machine Learning Repository using Pandas, and creates the feature matrix X and target vector y.

Splits the dataset into training and test sets using Scikit-Learn's train_test_split function.

Applies feature scaling to the training and test sets using Scikit-Learn's StandardScaler class.

Trains a logistic regression model on the training set using Scikit-Learn's LogisticRegression class.

Uses the trained model to make predictions on the test set.

Evaluates the performance of the model using a confusion matrix and accuracy score from Scikit-Learn's confusion_matrix and accuracy_score functions.

The output of the code will be the confusion matrix and the accuracy of the model on the test set.
